```python
import pandas as pd

import warnings
warnings.filterwarnings("ignore")

# surprise
!pip install scikit-surprise

from surprise import SVD
from surprise import Dataset
from surprise import accuracy
from surprise.model_selection import train_test_split
from surprise import Reader
```

# 데이터 불러오기

```python
traveler = pd.read_csv('/content/drive/MyDrive/3조(Rodi알려조)/9. 데이터셋/004.전처리된데이터/여행객(행분리).csv')
```

# 추천시스템 surprise

```python
# 평점의 범위를 지정(1점~3점)
reader = Reader(rating_scale=(1, 3))

# Surprise 데이터셋으로 변환(userID, itermID, rating)
data = Dataset.load_from_df(traveler[['MBTI', '선호여행', '만족도']], reader)

trainset, testset = train_test_split(data, test_size=0.2, random_state=0)
algo = SVD(random_state=0)
algo.fit(trainset)
predictions = algo.test(testset)
predictions[:5]
```

```python
accuracy.rmse(predictions)
```

```python
# 'ESFJ'인 사용자의 예측 결과를 저장할 리스트 생성
ESFJ_predictions = []

# predictions 리스트를 순회하면서 'ESFJ'인 사용자의 예측 결과를 esfj_predictions 리스트에 추가
for uid, iid, true_r, est, _ in predictions:
    if uid == 'ESFJ':
        ESFJ_predictions.append((uid, iid, true_r, est))

# esfj_predictions 리스트에서 첫 5개 결과를 출력
ESFJ_predictions.sort(reverse=True)
for prediction in ESFJ_predictions[:5]:
  print(prediction)
```

```python
# 'INTP'인 사용자의 예측 결과를 저장할 리스트 생성
INTP_predictions = []

# predictions 리스트를 순회하면서 'INTP'인 사용자의 예측 결과를 INTP_predictions 리스트에 추가
for uid, iid, true_r, est, _ in predictions:
    if uid == 'INTP':
        INTP_predictions.append((uid, iid, true_r, est))

# INTP_predictions 리스트에서 첫 5개 결과를 출력
INTP_predictions.sort(reverse=True)
for prediction in INTP_predictions[:5]:
  print(prediction)
```
