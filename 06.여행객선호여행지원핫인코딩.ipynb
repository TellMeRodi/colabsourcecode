# 선호여행지 원핫인코딩

선호여행지를 모두 컬럼으로 만든 후 순위에 따라 만족도를 해당 컬럼에 차등부여\
 \
예) \
a006685 선호여행지1 부산, 선호여행지2 대구, 선호여행지3 서울 ->\
a006685 선호여행지부산 3 선호여행지대구 2 선호여행지서울 1

```python
traveler1 = traveler[['TRAVELER_ID', '선호여행1']]
traveler1.rename(columns={'선호여행1':'선호여행'}, inplace=True)
traveler2 = traveler[['TRAVELER_ID', '선호여행2']]
traveler2.rename(columns={'선호여행2':'선호여행'}, inplace=True)
traveler3 = traveler[['TRAVELER_ID', '선호여행3']]
traveler3.rename(columns={'선호여행3':'선호여행'}, inplace=True)

traveler1 = pd.get_dummies(traveler1, columns=['선호여행'])
traveler1.replace(1, 3, inplace=True)
traveler2 = pd.get_dummies(traveler2, columns=['선호여행'])
traveler2.replace(1, 2, inplace=True)
traveler3 = pd.get_dummies(traveler3, columns=['선호여행'])

traveler = pd.concat([traveler, traveler1, traveler2, traveler3], axis=1, ignore_index=True)
```
